```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
quality=pd.read_csv('quality.csv')
quality.head()
```

```python
quality.tail()
```

```python
quality.info()
```

```python
quality.isnull().sum()
```

```python
 quality=quality.drop('type',axis=1)
```

```python
from sklearn.impute import SimpleImputer

imputer=SimpleImputer(strategy='median')
imputer.fit(quality)
```

```python
imputer.statistics_
```

```python
X=imputer.transform(quality)
quality=pd.DataFrame(X,columns=quality.columns)
```

```python
quality.describe()
```

```python
%matplotlib inline
quality.hist(bins=50,figsize=(12,10))
```

```python
from sklearn.model_selection import train_test_split
train_set,test_set=train_test_split(quality,test_size=0.2,random_state=42)
print(f"{train_set.shape} \n{test_set.shape}")
```

```python
quality=train_set.copy()
```

## Correlation

```python
corr_matrix=quality.corr()
```

```python
corr_matrix['quality'].sort_values()
```

```python
corr_matrix
```

```python
from pandas.plotting import scatter_matrix
attributes=['fixed acidity','volatile acidity','alcohol','density','chlorides','citric acid']
scatter_matrix(quality[attributes],figsize=(12,8))
```

```python
quality=train_set.drop('quality',axis=1)
quality_labels=train_set['quality'].copy()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(quality,quality_labels)
```

```python
some_data=quality.iloc[:5]
some_labels=quality_labels.iloc[:5]
```

```python
model.predict(some_data)
```

```python
some_labels
```

```python
from sklearn.metrics import mean_squared_error
Quality_prediction=model.predict(quality)
mse=mean_squared_error(quality_labels,Quality_prediction)
rms=np.sqrt(mse)
```

```python
rms
```

# Save 

```python
from joblib import dump,load
dump(model,'Wine Quality detection.joblib')
```

```python
X_test=test_set.drop('quality',axis=1)
Y_test=test_set['quality'].copy()

final_predictions=model.predict(X_test)
final_mse=mean_squared_error(Y_test,final_predictions)
final_rmse=np.sqrt(final_mse)
print(final_predictions)
```
